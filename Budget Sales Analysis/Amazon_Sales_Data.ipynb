```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x=data['Region'],hue=data['Region'])
plt.title('Region wise Sales Count')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x=data['Item Type'],y=data['Total Revenue'],hue=data['Item Type'])
plt.title('Total Revenue Generated by Each Item Type')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.scatterplot(x=data['Item Type'],y=data['Units Sold'],hue=data['Sales Channel'])
plt.title('Item Type Sales By Sales Channel')
plt.show()
```
